```python
import re
import matplotlib.pyplot as plt

log_file = "rl_training_20251203_231937.log"

steps = []
objectives = []

with open(log_file, "r") as f:
    curr_step = None
    for line in f:
        m = re.search(r"\[STEP (\d+)\]", line)
        if m:
            curr_step = int(m.group(1))
        m2 = re.search(r"OBJECTIVE: ([\d\.]+)", line)
        if m2 and curr_step is not None:
            steps.append(curr_step)
            objectives.append(float(m2.group(1)))

# EWMA implementation matching pandas with 'span' = window
def ewma(data, window):
    alpha = 2 / (window + 1)
    ewma_out = []
    for i, val in enumerate(data):
        if i == 0:
            ewma_out.append(val)
        else:
            ewma_out.append(alpha * val + (1 - alpha) * ewma_out[-1])
    return ewma_out

objectives_raw = objectives
ewma_window = 10
objectives_ewma_1 = ewma(objectives, window=1)  # This essentially returns the raw values
objectives_ewma_3 = ewma(objectives, window=ewma_window)

# Calculate EWMA for entire dataset, but plot only up to step 140
max_step_to_plot = 5000

plot_indices = [i for i, step in enumerate(steps) if step <= max_step_to_plot]

steps_plot = [steps[i] for i in plot_indices]
objectives_ewma_3_plot = [objectives_ewma_3[i] for i in plot_indices]

plt.figure(figsize=(8,5))
#plt.plot(steps, objectives_raw, marker="o", linestyle="--", color="tab:blue", alpha=0.4, label="Raw (EWMA=1)")
plt.plot(steps_plot, objectives_ewma_3_plot, linestyle="-", color="blue") #abel=f"EWMA (window={ewma_window})")
plt.title(f"Global Objective vs Step (WITHOUT OFFLINE TRAINING)")
plt.xlabel("Step No")
plt.ylabel("Global Objective")
plt.grid(True, linestyle="--", alpha=0.5)
plt.legend()
plt.tight_layout()

plt.savefig("global_objective_vs_step_without_offline_training.png")
plt.show()
```
